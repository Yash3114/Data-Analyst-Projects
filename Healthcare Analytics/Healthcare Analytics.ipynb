# Numerical Columns

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
numerical_cols = numerical_cols.drop('Room Number')
numerical_cols
```

```python
df[numerical_cols].describe()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

# Bivariate Analysis

```python
categorical_cols = ['Gender', 'Blood Type', 'Medical Condition', 'Admission Type', 'Insurance Provider', 'Medication', 'Test Results']

for col in categorical_cols:
    plt.figure(figsize=(8,5))
    group_data = df.groupby(col)['Billing Amount'].mean().sort_values()
    plt.plot(group_data.index, group_data.values, color='skyblue')
    plt.title(f'Average Billing Amount by {col}')
    plt.ylabel('Average Billing Amount')
    plt.xticks(rotation=45)
    plt.show()
```

```python
categorical_cols = ['Gender', 'Blood Type', 'Medical Condition', 'Admission Type', 'Insurance Provider', 'Medication', 'Test Results']

for col in categorical_cols:
    plt.figure(figsize=(8,5))
    group_data = df.groupby(col)['Age'].mean().sort_values()
    plt.plot(group_data.index, group_data.values, color = 'skyblue')
    plt.title(f'Average Age by {col}')
    plt.ylabel('Average Age')
    plt.xticks(rotation=45)
    plt.show()
```
